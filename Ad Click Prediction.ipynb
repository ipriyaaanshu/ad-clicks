# Ad Clicks Prediction

In this project I will be working with a fake advertising data set, indicating whether or not a particular internet user will click on an Advertisement. I will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Logistic Regression

Now it's time to do a train test split, and train the model!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income',
       'Daily Internet Usage', 'Male']]

y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 101)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost

logreg = LogisticRegression()
btc    = GradientBoostingClassifier()
dt     = DecisionTreeClassifier()
rf     = RandomForestClassifier()
xgb    = xgboost.XGBClassifier()
adb    = AdaBoostClassifier()
```

```python
logreg.fit(X_train,y_train)
btc.fit(X_train,y_train)
dt.fit(X_train,y_train)
rf.fit(X_train,y_train)
xgb.fit(X_train,y_train)
adb.fit(X_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
pred_logreg = logreg.predict(X_test)
pred_btc = btc.predict(X_test)
pred_dt = dt.predict(X_test)
pred_rf = rf.predict(X_test)
pred_xgb = xgb.predict(X_test)
pred_adb = adb.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,pred_logreg))
print()
print(metrics.classification_report(y_test,pred_btc))
print()
print(metrics.classification_report(y_test,pred_dt))
print()
print(metrics.classification_report(y_test,pred_rf))
print()
print(metrics.classification_report(y_test,pred_xgb))
print()
print(metrics.classification_report(y_test,pred_adb))
print()
```

```python
print(metrics.confusion_matrix(y_test,pred_logreg))
print()
print(metrics.confusion_matrix(y_test,pred_btc))
print()
print(metrics.confusion_matrix(y_test,pred_dt))
print()
print(metrics.confusion_matrix(y_test,pred_rf))
print()
print(metrics.confusion_matrix(y_test,pred_xgb))
print()
print(metrics.confusion_matrix(y_test,pred_adb))
print()
```

```python
print(metrics.roc_auc_score(y_test,pred_logreg))
print()
print(metrics.roc_auc_score(y_test,pred_btc))
print()
print(metrics.roc_auc_score(y_test,pred_dt))
print()
print(metrics.roc_auc_score(y_test,pred_rf))
print()
print(metrics.roc_auc_score(y_test,pred_xgb))
print()
print(metrics.roc_auc_score(y_test,pred_adb))
print()
```
